Osservando questi grafici è possibile notare come il sistema si equilibri dopo circa 40 blocchi da 10000 steps ciascuno. Nei successivi quattro grafici sono invece mostrate le simulazioni delle precedenti quantità termodinamiche al variare della temperatura, in un range di 0.5 - 2 K: le misure sono state prese in caso di sistema all'equilibrio termico, ovvero dopo il 40° blocco. Nei grafici sono messi a confronto le analisi prese coi due diversi metodi di campionamento (gibbs e metropolis) con le espressioni analitiche delle stesse quantità

```python
file=open("6_1/dati/output.ene.temp.metro")
sum=0
for line in file:
    sum+=1
file.close()

xM, meanETM, errorETM = np.loadtxt("6_1/dati/output.ene.temp.metro", usecols=(0,2,3), unpack='true')
xG, meanETG, errorETG = np.loadtxt("6_1/dati/output.ene.temp.gibbs", usecols=(0,2,3), unpack='true')

points=sum
T = np.linspace(0.5,2.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

plt.figure(figsize=(9,6))
plt.title('Ising 1D, internal energy', fontsize =25)
plt.xlabel('T')
plt.ylabel('U/N')

plt.plot(T, e)
plt.errorbar(xM,meanETM,yerr=errorETM)
plt.errorbar(xG,meanETG,yerr=errorETG)

blue_patch = mpatches.Patch(color='blue', label='Analitic expression (U)')
orange_patch = mpatches.Patch(color='orange', label='Metropolis sampling')
green_patch = mpatches.Patch(color='green', label='Gibbs sampling')
plt.legend(handles=[blue_patch,orange_patch,green_patch])

plt.grid(True)
plt.show()
```

```python
xM, meanCTM, errorCTM = np.loadtxt("6_1/dati/output.chi.temp.metro", usecols=(0,2,3), unpack='true')
xG, meanCTG, errorCTG = np.loadtxt("6_1/dati/output.chi.temp.gibbs", usecols=(0,2,3), unpack='true')

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

plt.figure(figsize=(9,6))
plt.plot(T, X)
plt.errorbar(xM,meanCTM,yerr=errorCTM)
plt.errorbar(xG,meanCTG,yerr=errorCTG)

plt.title('Ising 1D, Susceptibility', fontsize =25)
plt.xlabel('T')
plt.ylabel('$\chi$')

blue_patch = mpatches.Patch(color='blue', label='Analitic expression ($\chi$)')
orange_patch = mpatches.Patch(color='orange', label='Metropolis sampling')
green_patch = mpatches.Patch(color='green', label='Gibbs sampling')
plt.legend(handles=[blue_patch,orange_patch,green_patch])

plt.grid(True)
plt.show()
```

```python
xM, meanMTM, errorMTM = np.loadtxt("6_1/dati/output.mag.002temp.metro", usecols=(0,2,3), unpack='true')
xG, meanMTG, errorMTG = np.loadtxt("6_1/dati/output.mag.002temp.gibbs", usecols=(0,2,3), unpack='true')

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)


plt.figure(figsize=(9,6))
plt.plot(T, M)
plt.errorbar(xM,meanMTM,yerr=errorMTM)
plt.errorbar(xG,meanMTG,yerr=errorMTG)
plt.title('Ising 1D, magnetization M with h = 0.02', fontsize =25)
plt.xlabel('T')
plt.ylabel('$M$')

blue_patch = mpatches.Patch(color='blue', label='Analitic expression (M)')
orange_patch = mpatches.Patch(color='orange', label='Metropolis sampling')
green_patch = mpatches.Patch(color='green', label='Gibbs sampling')
plt.legend(handles=[blue_patch,orange_patch,green_patch])

plt.grid(True)
plt.show()
```
