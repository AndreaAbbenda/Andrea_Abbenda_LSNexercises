I due grafici seguenti mostrano gli scatterplot dei campionamenti di numeri casuali distribuiti secondo le distribuzioni di probabilit√† date dalle 2 funzioni d'onda di singola particella nei due autostati a n=1 e n=2, con n numero quantico fondamentale

```python
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

file=open("5_1/es5_1_N1.dat")
sum=0
for line in file:
    sum+=1
file.close()

X, Y, Z = np.loadtxt("5_1/es5_1_N1.dat", usecols=(0,1,2), unpack='true')

fig = plt.figure(figsize=(15,10))
fig.suptitle('Campionamento FEW - n=1', fontsize=25)
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x', fontsize=25)
ax.set_ylabel('y', fontsize=25)
ax.set_zlabel('z', fontsize=25)
ax.view_init(10, 30)
plt.show()
```

```python
file=open("5_1/es5_1_N2.dat")
sum=0
for line in file:
    sum+=1
file.close()

X2, Y2, Z2 = np.loadtxt("5_1/es5_1_N2.dat", usecols=(0,1,2), unpack='true')

plt.figure(figsize=(15,10))

fig = plt.figure(figsize=(15,10))
fig.suptitle('Campionamento FEW - n=2', fontsize=25)
ax = Axes3D(fig)
ax.scatter(X2, Y2, Z2, c=Z2, marker='.')
ax.set_xlabel('x', fontsize=25)
ax.set_ylabel('y', fontsize=25)
ax.set_zlabel('z', fontsize=25)
ax.view_init(10, 30)
plt.show()
```

Analizziamo ora la distanza media dal centro delle particelle campionate per entrambe le funzioni d'onda. In questo caso la probabilit√† di transizione 'tentativo' T(x|y) √® distibuita uniformemente

```python
import matplotlib.patches as mpatches

file=open("5_1/es5_1_R1.dat")
sum=0
for line in file:
    sum+=1
file.close()

mean1, error1 = np.loadtxt("5_1/es5_1_R1.dat", usecols=(0,1), unpack='true')
mean2, error2 = np.loadtxt("5_1/es5_1_R2.dat", usecols=(0,1), unpack='true')

plt.figure(figsize=(10,10))
plt.suptitle('Analisi <r>, T(x|y) distribuita uniformemente', fontsize=25)
x=np.arange(sum)
plt.errorbar(x,mean1,yerr=error1)
plt.errorbar(x,mean2,yerr=error2)
plt.xlabel('#blocks', fontsize =25)
plt.ylabel('<r>',fontsize=25)
plt.grid(True)
blue_patch = mpatches.Patch(color='blue', label='Fondamental quantic number n = 1')
orange_patch = mpatches.Patch(color='orange', label='Fondamental quantic number n = 2')
plt.legend(handles=[blue_patch,orange_patch])
plt.show()
print('Il valor medio di r nel caso di n = 1 √®: ', mean1[sum-1], '+-', error1[sum-1],' m/a0')
print('Il valor medio di r nel caso di n = 2 √®: ', mean2[sum-1], '+-', error2[sum-1],' m/a0')
```

Prendendo invece una T(x|y) distribuita normalmente otteniamo:

```python
file=open("5_1/es5_1_R1.dat")
sum=0
for line in file:
    sum+=1
file.close()

mean1_gauss, error1_gauss = np.loadtxt("5_1/es5_1_R1_gauss.dat", usecols=(0,1), unpack='true')
mean2_gauss, error2_gauss = np.loadtxt("5_1/es5_1_R2_gauss.dat", usecols=(0,1), unpack='true')

plt.figure(figsize=(10,10))
plt.suptitle('Analisi <r>, T(x|y) distribuita normalmente', fontsize=25)
x=np.arange(sum)
plt.errorbar(x,mean1_gauss,yerr=error1_gauss)
plt.errorbar(x,mean2_gauss,yerr=error2_gauss)
plt.xlabel('#blocks', fontsize=25)
plt.ylabel('<r>', fontsize=25)
plt.grid(True)
blue_patch = mpatches.Patch(color='blue', label='Fondamental quantic number n = 1')
orange_patch = mpatches.Patch(color='orange', label='Fondamental quantic number n = 2')
plt.legend(handles=[blue_patch,orange_patch])
plt.show()
print('Il valor medio di r nel caso di n = 1 √®: ', mean1[sum-1], '+-', error1[sum-1],' m/a0')
print('Il valor medio di r nel caso di n = 2 √®: ', mean2[sum-1], '+-', error2[sum-1],' m/a0')
```

Si pu√≤ quindi notare come variando la probabilit√† di transizione da un punto all'altro del random walk varia la distribuzione stessa dei punti nello spazio e di conseguenza il valor medio dell'osservabile r

Di seguito sono mostrati i due scatterplot e l'analisi del raggio medio dei precedenti sistemi dove per√≤ il punto iniziale del random walk √® preso molto distante dal centro (in questo caso nel punto O(20,20,20) inunit√† di a0). Si pu√≤ notare come, anche prendendo il punto O molto distante dall'origine del sistemadi riferimento, l'algoritmo funziona in egual modo ed in egual tempo, restituendo gli stessi valori medi dell'osservabile r

```python
file=open("5_1/es5_1_N1_20.dat")
sum=0
for line in file:
    sum+=1
file.close()

X1_20, Y1_20, Z1_20 = np.loadtxt("5_1/es5_1_N1_20.dat", usecols=(0,1,2), unpack='true')

fig = plt.figure(figsize=(15,10))
fig.suptitle('Campionamento FEW, n=1, O(20,20,20)', fontsize=25)
ax = Axes3D(fig)
ax.scatter(X1_20, Y1_20, Z1_20, c=Z1_20, marker='.')
ax.set_xlabel('x', fontsize =10)
ax.set_ylabel('y', fontsize=10)
ax.set_zlabel('z',fontsize=10)
ax.view_init(10, 30)
plt.show()
```

```python
file=open("5_1/es5_1_N2_20.dat")
sum=0
for line in file:
    sum+=1
file.close()

X2_20, Y2_20, Z2_20 = np.loadtxt("5_1/es5_1_N2_20.dat", usecols=(0,1,2), unpack='true')

fig = plt.figure(figsize=(15,10))
fig.suptitle('Campionamento FEW, n=2, O(20,20,20)', fontsize=25)
ax = Axes3D(fig)
ax.scatter(X2_20, Y2_20, Z2_20, c=Z2_20, marker='.')
ax.set_xlabel('x', fontsize =25)
ax.set_ylabel('y', fontsize=25)
ax.set_zlabel('z',fontsize=25)
ax.view_init(10, 30)
plt.show()
```

```python
file=open("5_1/es5_1_R1_20.dat")
sum=0
for line in file:
    sum+=1
file.close()

mean1_20, error1_20 = np.loadtxt("5_1/es5_1_R1_20.dat", usecols=(0,1), unpack='true')
mean2_20, error2_20 = np.loadtxt("5_1/es5_1_R2_20.dat", usecols=(0,1), unpack='true')

plt.figure(figsize=(10,10))
plt.suptitle('Analisi <r>, O(20,20,20)', fontsize=25)
x=np.arange(sum)
plt.errorbar(x,mean1_20,yerr=error1_20)
plt.errorbar(x,mean2_20,yerr=error2_20)
plt.xlabel('#blocks', fontsize=25)
plt.ylabel('<r>', fontsize=25)
plt.grid(True)
blue_patch = mpatches.Patch(color='blue', label='Fondamental quantic number n = 1')
orange_patch = mpatches.Patch(color='orange', label='Fondamental quantic number n = 2')
plt.legend(handles=[blue_patch,orange_patch])
plt.show()
print('Il valor medio di r nel caso di n = 1 √®: ', mean1_20[sum-1], '+-', error1_20[sum-1],' m/a0')
print('Il valor medio di r nel caso di n = 2 √®: ', mean2_20[sum-1], '+-', error2_20[sum-1],' m/a0')
```
