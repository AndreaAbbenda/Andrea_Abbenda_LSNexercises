### 4.1

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

file=open("4_1/es4_1_thermal.dat")
sum=0
for line in file:
    sum+=1
file.close()

temp = np.loadtxt("4_1/es4_1_thermal.dat", unpack='true')

plt.figure(figsize=(15,10))

x=np.arange(sum)
plt.plot(x,temp, 'b.')
plt.xlabel('step')
plt.ylabel('Temperature [U.L.J]')
plt.grid(True)
plt.show()
```

Nel grafico sopra vengono mostrati i valori di tempertura espressi in unità di lennard-jones nel caso di un sistema solido a 108 particelle interagenti con un potenziale di tipo lennard-jones. Viene evidenziato come, dato il valore scelto per la temperatura iniziale di circa 1.4 u.l.j., essa scenda molto dopo appena pochi steps ed è necessario che la procedura di inizializzazione venga ripetuta più volte (in questo caso 4) affinché la temperatura si assesti sul valore desisderato.
Ad ogni ripartenza il sistema evolve di 1000 step

### 4.2

Analisi dell'errore tramite metodo del data-blocking: 
energia totale di un solido a temperatura 1.4 u.l.j.

```python
file=open("4_2/output_err_etot.dat")
sum=0
for line in file:
    sum+=1
file.close()

mean, error = np.loadtxt("4_2/output_err_etot.dat", usecols=(0,1), unpack='true')

plt.figure(figsize=(10,10))

x=np.arange(sum)
plt.errorbar(x,mean,yerr=error)

plt.xlabel('#blocks')
plt.ylabel('Etot')
plt.grid(True)

plt.show()
```

## 4.3

```python
file=open("4_3/phases/output_err_solid_etot.dat")
sum=0
for line in file:
    sum+=1
file.close()

meanS, errorS = np.loadtxt("4_3/phases/output_err_solid_etot.dat", usecols=(0,1), unpack='true')

file=open("4_3/phases/output_err_liquid_etot.dat")
sum=0
for line in file:
    sum+=1
file.close()

meanL, errorL = np.loadtxt("4_3/phases/output_err_liquid_etot.dat", usecols=(0,1), unpack='true')

file=open("4_3/phases/output_err_gas_etot.dat")
sum=0
for line in file:
    sum+=1
file.close()

meanG, errorG = np.loadtxt("4_3/phases/output_err_gas_etot.dat", usecols=(0,1), unpack='true')

plt.figure(figsize=(15,10))

x=np.arange(sum)
plt.errorbar(x,meanS,yerr=errorS)
plt.errorbar(x,meanL,yerr=errorL)
plt.errorbar(x,meanG,yerr=errorG)

plt.xlabel('#blocks')
plt.ylabel('Temp [K]')
plt.grid(True)
blue_patch = mpatches.Patch(color='blue', label='Solid phase')
orange_patch = mpatches.Patch(color='orange', label='Liquid phase')
green_patch = mpatches.Patch(color='green', label='Gas phase')
plt.legend(handles=[blue_patch,orange_patch,green_patch])
plt.show()
```
