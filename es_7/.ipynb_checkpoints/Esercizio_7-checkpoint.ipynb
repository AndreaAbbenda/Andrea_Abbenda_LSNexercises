



Nei successivi grafici sono mostrate le funzioni di autocorrelazione $\chi(t)$ per energia e pressione in ognuna delle tre fasi (ogni timestep corrisponde ad una mossa accettata dall'algoritmo)

```python
def AutoCorr(data,Tmax):
    lg = (int)(len(data))
    ac_data = np.zeros(Tmax)
    m = 0
    m1 = 0
    m2 = 0
    for k in range(lg):
        m1 += data[k]
        m2 += data[k]**2
    for i in range(Tmax):
        m_m = 0
        m = 0
        m_ = 0
        for j in range(lg-i):
            m_m += data[j]*data[j+i]
            m += data[j]
            m_ += data[j+i]
        ac_data[i] = (m_m/(lg-i)-(m/(lg-i))*(m_/(lg-i)))/(m2/lg-(m1/lg)**2)
    return ac_data
```

```python
ac_Egas = AutoCorr(Egas,500)
```

```python
ac_Eliq = AutoCorr(Eliquid,500)
```

```python
ac_Esol = AutoCorr(Esolid,500)
```

```python
ac_Psol = AutoCorr(Psolid,500)
```

```python
ac_Pliq = AutoCorr(Pliquid,500)
```

```python
ac_Pgas = AutoCorr(Pgas,500)
```

```python
x_ac = np.arange(500)

fig, axs= plt.subplots(2,3,figsize=(15,10))
axs[0,0].plot(x_ac, ac_Esol)
axs[0,1].plot(x_ac, ac_Eliq)
axs[0,2].plot(x_ac, ac_Egas)
axs[1,0].plot(x_ac, ac_Psol)
axs[1,1].plot(x_ac, ac_Pliq)
axs[1,2].plot(x_ac, ac_Pgas)
axs[0, 0].set_title('Autocorrelazione: U/N del solido')
axs[0, 1].set_title('Autocorrelazione: U/N del liquido')
axs[0, 2].set_title('Autocorrelazione: U/N del gas')
axs[1, 0].set_title('Autocorrelazione: P del solido')
axs[1, 1].set_title('Autocorrelazione: P del liquido')
axs[1, 2].set_title('Autocorrelazione: P del gas')
axs[0, 0].set(xlabel='Time [t]', ylabel='$\chi(t)$')
axs[0, 1].set(xlabel='Time [t]', ylabel='$\chi(t)$')
axs[0, 2].set(xlabel='Time [t]', ylabel='$\chi(t)$')
axs[1, 0].set(xlabel='Time [t]', ylabel='$\chi(t)$')
axs[1, 1].set(xlabel='Time [t]', ylabel='$\chi(t)$')
axs[1, 2].set(xlabel='Time [t]', ylabel='$\chi(t)$')
axs[0, 0].grid(True)
axs[0, 1].grid(True)
axs[0, 2].grid(True)
axs[1, 0].grid(True)
axs[1, 1].grid(True)
axs[1, 2].grid(True)

plt.show()
```

Dai grafici si nota chiaramente come le funzioni di autocorrelazione in questione decadono al pi dopo circa 200 steps Monte Carlo. E' possibile quindi supporre come, volendo analizzare le quantit termodinamiche medie col metodo del data-blocking, dei blocchi da circa 1000 steps MC possano essere sufficientemente grandi per poter supporre misure scorrelate. Un'analisi ulteriore per stabilire la lunghezza dei blocchi  sotto riportata: l'intero campionamento di 500000 steps MC viene scomposto in blocchi contenenti da 10 a 5000 campionamenti e per ogni ripartizione viene calcolata l'incertezza statistica  $\sigma$ col metodo del data-blocking

```python
def ErrVett(vett, num):
    lgt = len(vett)
    err = np.zeros(num)
    for i in range(num):
        
        steps_per_block = 10*(i+1)
        n_blocks = lgt//steps_per_block
        
        ave = np.zeros(n_blocks)
        ave2 = np.zeros(n_blocks)
        sum1 = 0
        sum2 = 0
        for j in range(n_blocks):
            for k in range(steps_per_block):
                ave[j] += vett[j*steps_per_block+k]
            ave[j] = ave[j]/steps_per_block
            ave2[j] = ave[j]**2
            sum1 += ave[j]
            sum2 += ave2[j]
        err[i] = math.sqrt((sum2/n_blocks - (sum1/n_blocks)**2)/(n_blocks-1))
    return err
```

```python
errES = ErrVett(Esolid,500)
```

```python
errEL = ErrVett(Eliquid,500)
```

```python
errEG = ErrVett(Egas,500)
```

```python
errPS = ErrVett(Psolid,500)
```

```python
errPL = ErrVett(Pliquid,500)
```

```python
errPG = ErrVett(Pgas,500)
```

```python
x_err = np.arange(500) * 10
mean0 = np.zeros(500)

fig_err, axs_err= plt.subplots(2,3,figsize=(15,10))

axs_err[0,0].errorbar(x_err, mean0, errES)
axs_err[0,1].errorbar(x_err, mean0, errEL)
axs_err[0,2].errorbar(x_err, mean0, errEG)
axs_err[1,0].errorbar(x_err, mean0, errPS)
axs_err[1,1].errorbar(x_err, mean0, errPS)
axs_err[1,2].errorbar(x_err, mean0, errPS)
axs_err[0, 0].set_title('Incertezza statistica: U/N del solido')
axs_err[0, 1].set_title('Incertezza statistica: U/N del liquido')
axs_err[0, 2].set_title('Incertezza statistica: U/N del gas')
axs_err[1, 0].set_title('Incertezza statistica: P del solido')
axs_err[1, 1].set_title('Incertezza statistica: P del liquido')
axs_err[1, 2].set_title('Incertezza statistica: P del gas')
axs_err[0, 0].set(xlabel='Steps per block', ylabel='$\sigma$')
axs_err[0, 1].set(xlabel='Steps per block', ylabel='$\sigma$')
axs_err[0, 2].set(xlabel='Steps per block', ylabel='$\sigma$')
axs_err[1, 0].set(xlabel='Steps per block', ylabel='$\sigma$')
axs_err[1, 1].set(xlabel='Steps per block', ylabel='$\sigma$')
axs_err[1, 2].set(xlabel='Steps per block', ylabel='$\sigma$')
axs_err[0, 0].grid(True)
axs_err[0, 1].grid(True)
axs_err[0, 2].grid(True)
axs_err[1, 0].grid(True)
axs_err[1, 1].grid(True)
axs_err[1, 2].grid(True)

plt.show()
```

Dai grafici si nota come la $\sigma$ saturi rapidamente:  possibile quindi stabilire che per il metodo del data-blocking siano sufficienti blocchi da 1000 campionamenti ciascuno, in quanto cos facendo le misure risultano sufficientemente scorrelate e si ottiene la massima incertezza statistica.
