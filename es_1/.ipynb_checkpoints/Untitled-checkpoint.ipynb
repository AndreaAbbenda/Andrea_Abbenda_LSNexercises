```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

file=open("1_1/es1_1_mean.txt")
sum=0
for line in file:
    sum+=1
file.close()

mean, error = np.loadtxt("1_1/es1_1_mean.txt", usecols=(0,1), unpack='true')

x=np.arange(sum)
plt.errorbar(x,mean,yerr=error)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
file=open("1_1/es1_1_err.txt")
sum=0
for line in file:
    sum+=1
file.close()

mean, error = np.loadtxt("1_1/es1_1_err.txt", usecols=(0,1), unpack='true')

x=np.arange(sum)
plt.errorbar(x,mean,yerr=error)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

### 1.2

```python
import math
from scipy.optimize import curve_fit
from scipy import stats

def gauss(x, mu, sigma):
    return ((1/(2*math.pi*sigma**2)) * np.exp(-1.0 * (x - mu)**2 / (2 * sigma**2)))
def lorentz(x, gamma, mean):
    return (1/(math.pi) * gamma/(gamma**2 + (x-mean)**2))

exp1, exp2, exp10, exp100 = np.loadtxt("1_2/es1_2_exp.txt", usecols=(0,1,2,3), unpack='true')
n_bins = 50

fig, axs= plt.subplots(2,2,figsize=(10,10))
axs[0,0].hist([exp1], n_bins, range=(-0.2,4), density=True, histtype='stepfilled')
axs[0,1].hist([exp2], n_bins, range=(-0.2,4), density=True, histtype='stepfilled')
axs[1,0].hist([exp10], n_bins, range=(-0.2,4), density=True, histtype='stepfilled')
axs[1,1].hist([exp100], n_bins, range=(-0.2,4), density=True, histtype='stepfilled')

plt.figure(figsize=(10,10))

n, bins, patches = plt.hist(exp100, n_bins, range=(0,2), density=True)
cut=bins[1:]
g_opt, g_cov = curve_fit(gauss, xdata=cut, ydata=n, bounds=([0.5, 0.1],[1.5, 0.5]))
l_opt, l_cov = curve_fit(lorentz, xdata=cut, ydata=n)

y_fit = gauss(cut,g_opt[0],g_opt[1])
plt.plot(cut,y_fit)

y_fit = lorentz(cut,l_opt[0],l_opt[1])
plt.plot(cut,y_fit)

print(g_opt)
```

```python
lor1, lor2, lor10, lor100 = np.loadtxt("1_2/es1_2_lor.txt", usecols=(0,1,2,3), unpack='true')
n_bins = 50

fig, axs= plt.subplots(2,2,figsize=(10,10))
axs[0,0].hist([lor1], n_bins, range=(-5,5), density=True, histtype='stepfilled')
axs[0,1].hist([lor2], n_bins, range=(-5,5), density=True, histtype='stepfilled')
axs[1,0].hist([lor10], n_bins, range=(-5,5), density=True, histtype='stepfilled')
axs[1,1].hist([lor100], n_bins, range=(-5,5), density=True, histtype='stepfilled')

plt.figure(figsize=(10,10))

n, bins, patches = plt.hist(lor100, n_bins, range=(-5,5), density=True)
cut=bins[1:]
g_opt, g_cov = curve_fit(gauss, xdata=cut, ydata=n)
l_opt, l_cov = curve_fit(lorentz, xdata=cut, ydata=n)

y_fit = gauss(cut,g_opt[0],g_opt[1])
plt.plot(cut,y_fit)

y_fit = lorentz(cut,l_opt[0],l_opt[1])
plt.plot(cut,y_fit)

print(g_opt)
```

```python
std1, std2, std10, std100 = np.loadtxt("1_2/es1_2_std.txt", usecols=(0,1,2,3), unpack='true')
n_bins = 50

fig, axs= plt.subplots(2,2,figsize=(10,10))
axs[0,0].hist([std1], n_bins, range=(0,1), density=True, histtype='stepfilled')
axs[0,1].hist([std2], n_bins, range=(0,1), density=True, histtype='stepfilled')
axs[1,0].hist([std10], n_bins, range=(0,1), density=True, histtype='stepfilled')
axs[1,1].hist([std100], n_bins, range=(0,1), density=True, histtype='stepfilled')

plt.figure(figsize=(10,10))

n, bins, patches = plt.hist(std100, n_bins, range=(0,1), density=True)
cut=bins[1:]
g_opt, g_cov = curve_fit(gauss, xdata=cut, ydata=n, bounds=([0.4,0.2],[0.6,0.3]))
l_opt, l_cov = curve_fit(lorentz, xdata=cut, ydata=n)

y_fit = gauss(cut,g_opt[0],g_opt[1])
plt.plot(cut,y_fit)

y_fit = lorentz(cut,l_opt[0],l_opt[1])
plt.plot(cut,y_fit)

print(g_opt)
```
