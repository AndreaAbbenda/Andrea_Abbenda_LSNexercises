- è stata modificata la funzione Mutate(): ogni evoluzione della Population comporta la selezione tramite la funzione Select di N/2 coppie di Chromosomes (con N = n° totale di Chromosomes per Population) e ogni coppia ha il 60% di probabilità di effettuare un crossover; successivamente ogni Chromosome ha il 10% di probabilità di andare incontro ad una delle 4 mutazioni possibili (Switch(), Shift(), Permute() o Invert()). Alla fine di questo processo viene calcolata la Fitness del nuovo Chromosome ed esso viene accettato con probabilità di accettazione $p_i=min[1,e^{-\frac{F_{new}-F_{old}}{T}}]$

Di seguito sono messi a paragone i risultati ottenuti con il semplice algoritmo genetico con quelli ottenuti tramite l'algoritmo del simulated annealing. Sono riportati i grafici della norma $L^1$ del miglior cammino, calcolati per entrambe le disposizioni delle città (disposizione casuale in un box di lato 100 e disposizione circolare su una circonferenza di lato 100) e gli scatterplot del miglior cammino calcolato dall'algoritmo

```python
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
xc, yc = np.loadtxt("10_1/30/SA/best1_circle.dat", usecols=(0,1), unpack='true')
xs, ys = np.loadtxt("10_1/30/SA/best1_square.dat", usecols=(0,1), unpack='true')
oxc, oyc = np.loadtxt("10_1/30/old/best1_circle.dat", usecols=(0,1), unpack='true')
oxs, oys = np.loadtxt("10_1/30/old/best1_square.dat", usecols=(0,1), unpack='true')


Lc = np.loadtxt("10_1/30/SA/L1_circle.dat", usecols=(0), unpack='true')
Ls = np.loadtxt("10_1/30/SA/L1_square.dat", usecols=(0),  unpack='true')

oLc = np.loadtxt("10_1/30/old/L1_circle.dat", usecols=(0), unpack='true')
oLs = np.loadtxt("10_1/30/old/L1_square.dat", usecols=(0),  unpack='true')


fig, ax = plt.subplots(2, 1, figsize=(10, 10))

ax[0].plot(Lc, label='Simulated annealing')
ax[0].plot(oLc, label='Standard mutations')
ax[0].set_title('Fitness del miglior cammino al variare delle mutazioni, disposizione circolare')
ax[0].set_xlabel('#mutations')
ax[0].set_ylabel('norma $L^1$')
ax[0].legend()
ax[0].grid(True)

ax[1].plot(Ls, label='Simulated annealing')
ax[1].plot(oLs, label='Standard mutations')
ax[1].set_title('Fitness del miglior cammino al variare delle mutazioni, disposizione casuale in un box quadrato')
ax[1].set_xlabel('#mutations')
ax[1].set_ylabel('norma $L^1$')
ax[1].legend()
ax[1].grid(True)


plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(15, 15))

ax[0,0].scatter(xc, yc)
ax[0,0].plot(xc, yc)
ax[0,0].set_title('Simulated annealing, disposizione circolare')

ax[0,1].scatter(oxc, oyc)
ax[0,1].plot(oxc, oyc)
ax[0,1].set_title('Standard genetic algoritm, disposizione circolare')

ax[1,0].scatter(xs, ys)
ax[1,0].plot(xs, ys)
ax[1,0].set_title('Simulated annealing, disposizione casuale in un box quadrato')

ax[1,1].scatter(oxs, oys)
ax[1,1].plot(oxs, oys)
ax[1,1].set_title('Standard genetic algoritm, disposizione casuale in un box quadrato')

fig.suptitle('Scatterplot dei migliori percorsi ottenuti dai due algoritmi',fontsize=15)

plt.show()
```

Parallelize with MPI libraries your Genetic Algorithm code in order to solve the TSP by performing a *Genetic Search* with 4 **parallel GA searches of the optimal path**:

E' stato parallelizzato il codice dell'algoritmo genetico ed è stato fatto girare su 4 'Continents'. Ogni Continent ha eseguito 30000 mutazioni ed ogni 2000 di esse i Continents si si sono scambiati i percorsi migliori. Di seguito sono messi a confronto i risultati ottenuti con l'algoritmo genetico parallelizzato con quelli ottenuti precedentemente con simulated annealing e algoritmo genetico semplice. 

```python
pxs, pys = np.loadtxt("10_2/results/best1_square.dat", usecols=(0,1), unpack='true')

pLs = np.loadtxt("10_2/results/L1_square_best.dat", usecols=(0),  unpack='true')

fig, ax = plt.subplots(2, 1, figsize=(8, 15))

ax[0].plot(pLs, label='Parallel genetic algorithm')
ax[0].plot(Ls, label='Simulated annealing')
ax[0].plot(oLs, label='Normal genetic algorithm')
ax[0].set_title('Fitness del miglior cammino al variare delle mutazioni, disposizione casuale in un box quadrato')
ax[0].set_xlabel('#mutations')
ax[0].set_ylabel('norma $L^1$')
ax[0].legend()
ax[0].grid(True)

ax[1].scatter(pxs, pys)
ax[1].plot(pxs, pys)
ax[1].set_title('Parallel genetic algorithm, disposizione casuale in un box quadrato')

plt.show()

print('Parallel genetic algorithm, best pattern:', pLs[len(pLs)-1])
print('Simulated annealing, best pattern:', Ls[len(Ls)-1])
print('Normal genetic algorithm, best pattern:', oLs[len(oLs)-1])
```

Dal primo grafico emerge chiaramente come parallelizzando il codice dell'algoritmo genetico si ottenga un netto miglioramento del miglior cammino, abbassandone quasi del 20% il valore rispetto ai precedenti metodi computazionali
