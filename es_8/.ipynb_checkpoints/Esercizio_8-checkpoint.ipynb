### Exercise 08.1

<span style="color:blue">

Write a C++ Variational Monte Carlo code for a single quantum particle in 1D which exploits the Metropolis algorithm to sample a trial wave function $|\Psi_T(x)|^2$ using an uniform transition probability $T(x_{new}|x_{old})$ (Suggestion: as a starting point, you can use the code you have already developed for the Hydrogen atom or the Monte Carlo NVT code). Using data blocking, the code should be able to compute the expectation value for the Hamiltonian

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

which can be used to find the parameters that minimize this quantity.
    
</span>

Il metodo variazionale per la stima di $\langle {\hat H} \rangle_T $ data una funzione d'onda di prova $\Psi_T(x)$ è stato implementato nel codice C++. Per farlo è stata creata una classe WaveFunction nel file random.cpp con datamembri _mu e _sigma nella quale sono stati implementati i seguenti metodi:

- $ Eval(x) = \Psi_T(x) = \frac{e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}}{\sqrt{2\sigma\sqrt{\pi}(1+e^{-\frac{\mu^2}{\sigma^2}})}}$


- $ Eval2(x) = |\Psi_T(x)|^2$


- $ D2Eval(x) = \frac{\partial^2}{\partial x^2}\Psi_T(x) =
\frac{\frac{x^2+\mu^2-\sigma^2}{\sigma^4}(e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}) + \frac{2x\mu}{\sigma^4}(e^{-\frac{(x+\mu)^2}{2\sigma^2}}-e^{-\frac{(x-\mu)^2}{2\sigma^2}})}{\sqrt{2\sigma\sqrt{\pi}(1+e^{-\frac{\mu^2}{\sigma^2}})}}$


- $ Ham(x) = -\frac{1}{2}\frac{\partial^2}{\partial x^2}\Psi_T(x) + (x^4 - \frac{5}{2}x^2)\Psi_T(x) $

Il codice campiona la funzione d'onda scelta sfuttando l'algoritmo di Metropolis, usando come probabilità di accettazione:

$$ A(x_{new},x_{old}) = min \left[ 1, |\Psi_T(x)|^2 \right] $$

<span style="color:blue">
Use your Variational Monte Carlo code to find the parameters $\sigma$ and $\mu$ which minimize $\langle {\hat H} \rangle_T$:

$$
\langle {\hat H} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}
{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

- As usual,show a picture of the estimation of $\langle {\hat H} \rangle_T$  and its statistical uncertainty as a function of the number of blocks/MC steps for the set of parameters which minimize $\langle {\hat H} \rangle_T$
- show also a picture **of the sampled** $|\Psi_T(x)|^2$ by filling a histogram with the sampled configurations, moreover compare it with the analytic curve of $|\Psi_T(x)|^2$ and with the numerical solution obtained by transforming the Schrodinger equation into a matrix equation (see below).
</span>

Il codice variazionale Monte Carlo (VMC) per la stima dei parametri $\mu$ e $\sigma$ che minimizzano $\langle {\hat H} \rangle_T$ è stato implementato sfuttando l'algoritmo di simulated annealing:
- data una coppia iniziale di valori $\mu_0 = 1$ e $\sigma_0 = 1$ è generata una seconda coppia di valori $\mu_{test}$ e $\sigma_{test}$ distribuiti uniformemente intorno ai precedenti
- vengono calcolate $\langle {\hat H} \rangle_0$ e $\langle {\hat H} \rangle_{test}$ utilizzando la funzione $AveH(x)$ della classe WaveFunction, nella quale è riportato in maniera pressoché identica il codice scritto nel punto 8.1, con la differenza che il calcolo è effettuato su un solo blocco da 5000 valori 
- sfruttando l'algoritmo di Metropolis la mossa proposta viene accettata con la seguente probabilità:
$$ A(\left[\mu_0,\sigma_0\right],\left[\mu_{new},\sigma_{old}\right]) = min \left[ 1, e^{-\frac{\langle {\hat H} \rangle_0 - \langle {\hat H} \rangle_{old}}{T_f}} \right] $$
- $T_f$ è una temperatura fittizia che viene fatta scendere da 1 a 0.02 e per ogni suo valore vengono proposte 1000 evoluzioni di $\mu$ e $\sigma$

Il valore finale di $\sigma_0$ e $\mu_0$ è quello che minimizza il funzionale:

$$\langle {\hat H_{\sigma,\mu}} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}
{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle}$$


Di seguito è riportato l'istogramma 2D dei campionamenti di $\sigma$ e $\mu$ accettati dall'algoritmo:

Nel grafico seguente sono messi a confronto l'istogramma dei campionamenti di $\Psi_T^{\tilde \sigma,\tilde \mu}$ con la funzione d'onda di ground state stimata numericamente discretizzando l'equazione di Schrodinger. E' inoltre mostrato sullo stesso grafico il potenziale $V(x)$

```python
xi, phi, H_p = np.loadtxt("8_2/best/sampling_best.dat", usecols=(0,1,2), unpack='true')


def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal


# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])


# Plot a few things
plt.figure(figsize=(15,7))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2, label='Numerical solution')
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-2.5,2.5))
plt.ylim((-0.6,0.6))
n, bins, patches = plt.hist(xi, 200, range=(-a/2,a/2), density=True, label='VMC')
plt.legend()
plt.show()
```

Viene infine riportato il grafico di $\langle {\hat H_{\tilde \sigma,\tilde \mu}} \rangle_T$ stimato col metodo del data-blocking al variare del numero di blocchi. Si nota come il valore dell'energia media stimata con il Variational Monte Carlo si avvicini al valore stimato numericamente (più accurato) senza esserne però compatibile: ciò è dovuto al fatto che la scelta iniziale della forma di $\Psi_T(x)$ non è corretta e non corrisponde alla funzione d'onda di GS reale della particella

```python
Hmean, Herror = np.loadtxt("8_2/best/hamilton_best.dat", usecols=(0,1), unpack='true')

plt.figure(figsize=(10,6))

Hx=np.arange(len(Hmean))
plt.errorbar(Hx,Hmean,yerr=Herror, label='$< {\hat H_{\hat \sigma,\hat \mu}} >_T$')
plt.plot(Hx, [-0.46046 for i in Hx],  label='exact ground state')

plt.xlabel('#blocks')
plt.ylabel('$< {\hat H} >_T $')
plt.title("Stima di $< {\hat H_{\hat \sigma,\hat \mu}} >_T$ per la funzione d'onda ottimale $\Psi_T^{ \hat\sigma,\hat \mu}(x)$",fontsize=20)
plt.grid(True)
plt.legend()

plt.show()
```
