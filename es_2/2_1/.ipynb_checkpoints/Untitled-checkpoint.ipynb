```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

file=open("es1_1_mean.txt")
sum=0
for line in file:
    sum+=1
file.close()

mean, error = np.loadtxt("es1_1_mean.txt", usecols=(0,1), unpack='true')

x=np.arange(sum)
plt.errorbar(x,mean,yerr=error)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
file=open("es1_1_err.txt")
sum=0
for line in file:
    sum+=1
file.close()

mean, error = np.loadtxt("es1_1_err.txt", usecols=(0,1), unpack='true')

x=np.arange(sum)
plt.errorbar(x,mean,yerr=error)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```
